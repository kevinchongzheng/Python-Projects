```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
from google.colab import files
```

```python
#Importing COVID-19 dataset
covid_19_csv = files.upload()
import io
df = pd.read_csv(io.BytesIO(covid_19_csv['covid19_Confirmed_dataset.csv']))

df.head()
```

```python
df.shape
```

```python
df.drop(['Lat','Long'],axis=1,inplace=True)
df
```

```python
df_aggregated = df.groupby('Country/Region').sum()
df_aggregated
```

```python
df_aggregated.shape
```

```python
df_aggregated.loc['China'].plot()
df_aggregated.loc['Italy'].plot()
df_aggregated.loc['Spain'].plot()
plt.pyplot.legend()
```

```python
#calculating and plotting the first derivative of the curve
df_aggregated.loc['China'].diff().plot()
```

Find max infection rate in China, Italy, Spain

```python

df_aggregated.loc['China'].diff().max()
```

```python
df_aggregated.loc['Italy'].diff().max()
```

```python
df_aggregated.loc['Spain'].diff().max()
```

Find max infection for all coutries

```python
countries = list(df_aggregated.index)
max_infection_rates = []
for c in countries:
  max_infection_rates.append(df_aggregated.loc[c].diff().max())
df_aggregated['max_infection_rate'] = max_infection_rates

df_aggregated.head()
```

Create a new dataframe with max_infection_rate column

```python
max_infection_countries = pd.DataFrame(df_aggregated['max_infection_rate'])
max_infection_countries
```

```python
happiness_report = files.upload()
df_happiness = pd.read_csv(io.BytesIO(happiness_report['worldwide_happiness_report.csv']))
```

```python
df_happiness.head()
```

```python
useless_cols = ['Overall rank','Score','Generosity','Perceptions of corruption']
df_happiness.drop(useless_cols,axis=1,inplace=True)
```

```python
df_happiness.head()
```

```python
df_happiness.set_index('Country or region',inplace=True)
```

```python
df_happiness.head()
```

Join two dataset

```python
data = max_infection_countries.join(df_happiness, how='inner')
```
